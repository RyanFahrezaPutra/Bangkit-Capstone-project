```python
from keras.preprocessing.image import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
from google.colab import drive

drive.mount('/content/drive')
```

```python
model = tf.keras.models.load_model('/content/drive/Shareddrives/C241-PS501/ML/fish_disease.h5')
```

```python
class_names = np.load('/content/drive/Shareddrives/C241-PS501/ML/nama_kelas.npy', allow_pickle=True)
```

```python
from google.colab import files

uploaded = files.upload()
img_path = list(uploaded.keys())[0]
img = load_img(img_path, target_size=(299, 299))

plt.imshow(img)

x = img_to_array(img) / 255.
x = np.expand_dims(x, axis=0)
prediction = model.predict(x)
index = np.argmax(prediction)
plt.axis('off')
plt.text(0, -20, 'Prediksi: {} {:.2f}%'.format(class_names[index], prediction[0][index]*100))
plt.show()

for i in range(len(class_names)):
    print(f"{class_names[i]}: {prediction[0][i]*100 :.2f}% ")
```
